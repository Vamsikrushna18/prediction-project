```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# Select only numeric columns (excluding non-numeric columns like 'name' and 'status')
numeric_columns = parkinsons_dataset.select_dtypes(include=['number']).columns

# Group by 'status' and calculate the mean for the numeric columns only
grouped_mean = parkinsons_dataset[numeric_columns].groupby(parkinsons_dataset['status']).mean()

# Display the result
print(grouped_mean)
```

```python
x = parkinsons_dataset.drop(columns=['status'],axis=1)
y = parkinsons_dataset['status']
```

```python
x.head()
```

```python
X_train , X_test , Y_train ,Y_test =train_test_split (x,y,test_size = 0.2 , random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
model = svm.SVC(kernel='linear')
```

```python
# Drop non-numeric columns from X_train (e.g., the 'name' column or any other string column)
X_train_numeric = X_train.select_dtypes(include=['number'])

# Ensure Y_train is numeric (it should be, but this is to double-check)
Y_train_numeric = Y_train.astype('int')  # or 'float' depending on your dataset

# Fit the model using only numeric features
model.fit(X_train_numeric, Y_train_numeric)
```

```python


# Drop the 'name' column (or any other non-numeric column) from both training and test sets
X_train_clean = X_train.drop(columns=['name'], axis=1)
X_test_clean = X_test.drop(columns=['name'], axis=1)

# Ensure that both train and test have the same columns in the same order
X_train_clean, X_test_clean = X_train_clean.align(X_test_clean, join='inner', axis=1)

# Now, you can fit the model again
model.fit(X_train_clean, Y_train)

# And make predictions
y_pred = model.predict(X_test_clean)
```

```python
y_pred = model.predict(X_test_clean)
```

```python
print(accuracy_score(Y_test,y_pred))
```
